# Implementation with Keras
ImageDataGenerator Generate batches of tensor image data with real-time data augmentation.


[Source - Data Augmentation | How to use Deep Learning when you have Limited Data](https://nanonets.com/blog/data-augmentation-how-to-use-deep-learning-when-you-have-limited-data-part-2/)

```python
import matplotlib.pyplot as plt
from numpy import expand_dims
from keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
img = load_img("/content/drive/MyDrive/Colab Notebooks/Augmentation /cat.jpg")
plt.imshow(img)
```

## Function that will show sub plots of the results of our augmentation

```python
def sub_plot_maker(datagen, plots, size):
  # batch of 3 images (mul the cat image into 3 images)
  _ = datagen.flow(samples, batch_size=1) 
  for i in range(plots):
    batch = _.next()
    image = batch[0].astype('uint8')
    plt.subplot(size + 1 + i)
    plt.imshow(image)
    plt.xticks([])
    plt.yticks([])
  plt.show()
```

We expand the dim that we can use an array of images

```python
data = img_to_array(img)
print(f'dim befor expand {data.shape}')

samples = expand_dims(data, 0) 
print(f'dim after expand {samples.shape}')
```

## Hoizontal shift

```python
datagen = ImageDataGenerator(width_shift_range=[-50,50])
sub_plot_maker(datagen, 3, 130)
```

## Vertical shift

```python
datagen = ImageDataGenerator(height_shift_range=[-50,50])
sub_plot_maker(datagen, 3, 130)
```

## Horizontal *flipping*

```python
datagen = ImageDataGenerator(horizontal_flip=True)
sub_plot_maker(datagen, 3, 130)
```

## Vertical flipping

```python
datagen = ImageDataGenerator(vertical_flip=True)
sub_plot_maker(datagen, 3, 130)
```

## Rotation

```python
datagen = ImageDataGenerator(rotation_range=90)
sub_plot_maker(datagen, 3, 130)
```

```python
datagen = ImageDataGenerator(zoom_range=[0.1,2])
sub_plot_maker(datagen, 3, 130)
```

## Brightness

```python
datagen = ImageDataGenerator(brightness_range=[0.5,2])
sub_plot_maker(datagen, 3, 130)
```

##Shear

```python
datagen = ImageDataGenerator(shear_range=45)
sub_plot_maker(datagen, 3, 130)
```

## Channel Shift
For example, Color image has 3 channels R,B, and G 

```python
datagen = ImageDataGenerator(channel_shift_range=90)
sub_plot_maker(datagen, 3, 130)
```

# Multi Augmentation Generator

```python
datagen = ImageDataGenerator(channel_shift_range=180,
                             brightness_range=[0.5,2],
                             height_shift_range=[-50,50],
                             horizontal_flip=True,
                             rotation_range=90)
sub_plot_maker(datagen, 9, 330)
```
